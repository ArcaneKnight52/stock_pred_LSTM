```python
from sklearn.model_selection import  train_test_split
from sklearn.preprocessing import LabelEncoder
#it is used to splite which data is to train and test 
# it is used to encode the each by like maping
```

```python
le = LabelEncoder()
df['CROP TYPE'] = le.fit_transform(df['CROP TYPE'])
df['SOIL TYPE'] = le.fit_transform(df['SOIL TYPE'])
df['REGION'] = le.fit_transform(df['REGION'])
df['TEMPERATURE'] = le.fit_transform(df['TEMPERATURE'])
df['WEATHER CONDITION'] = le.fit_transform(df['WEATHER CONDITION'])
#it is to encode the data from the sheet to train
```

```python
x = df[['CROP TYPE','SOIL TYPE','REGION','TEMPERATURE','WEATHER CONDITION']]
y = df[['WATER REQUIREMENT']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
#it will train and test the value agt ratio of .2
```

```python
lr.fit(x_train,y_train)
# it will fit the data with the module and train
```

```python
y_pred = lr.predict(x_test)
#it will predict the value
```
